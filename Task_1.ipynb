```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_url = "https://datahub.io/core/population/r/population.csv"
df = pd.read_csv(file_url)

print(df.head())


df_filtered = df[df["Year"] == 2020]

# Plot histogram for population distribution
plt.figure(figsize=(10, 5))
sns.histplot(df_filtered["Value"], bins=30, kde=True, color='blue')
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title("Population Distribution (Year 2020)")
plt.show()

# Bar chart for top 10 most populated countries in 2020
top_10_countries = df_filtered.nlargest(10, "Value")

plt.figure(figsize=(12, 6))
sns.barplot(x="Country Name", y="Value", data=top_10_countries, palette="viridis")
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Most Populated Countries (2020)")
plt.xticks(rotation=45)
plt.show()
```
